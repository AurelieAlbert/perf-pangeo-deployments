```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib 

matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
matplotlib.rcParams.update({'font.size': 20})
```

```python
workers2=[8,30,120,200,368]
workers4=[80,240,736]
workers8=[32,200,1470]
timing2=[45,48,45,48,45]
timing_error2=[0,0,0,0,0]
timing4=[24,22,23]
timing_error4=[0,0,0]
timing8=[12,11.6,15]
timing_error8=[0,0,0]
timing_scalable2=[45,45*8/30,45*8/120,45*8/200,45*8/368]
timing_scalable4=[24,24*80/240,24*80/736]
timing_scalable8=[12,12*32/200,12*32/1470]
```

```python
marker_style = dict(linestyle=':', color='cornflowerblue', markersize=10)

fig, ax = plt.subplots()
fig.figsize=(20, 8)
plt.errorbar(workers2,timing2,yerr=timing_error2,fmt='o-',color='cornflowerblue',label='results 2 workers')
plt.plot(workers2,timing_scalable2,'.-',marker='X',color='black',label='scalable 2 workers')
plt.errorbar(workers4,timing4,yerr=timing_error4,fmt='o:',color='cornflowerblue',label='results 4 workers')
plt.plot(workers4,timing_scalable4,'.:',marker='X',color='black',label='scalable 4 workers')
plt.errorbar(workers8,timing8,yerr=timing_error8,fmt='o--',color='cornflowerblue',label='results 8 workers')
plt.plot(workers8,timing_scalable8,'.--',marker='X',color='black',label='scalable 8 workers')


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Number of workers')
plt.ylabel('Walltime in minutes')
plt.title('Performance tests of HAL - fixed number of workers/cores')
plt.savefig('figs/Perf-hal2.png', bbox_inches='tight')
```
